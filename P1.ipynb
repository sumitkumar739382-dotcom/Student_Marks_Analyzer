Student Marks Analyzer

Input: Student names, marks in subjects.

Output: Average, highest, lowest marks, and subject-wise topper.

Concepts: Loops, dictionaries, list manipulation.

```python
import pandas as pd
import statistics as st
#Marks analysis function
def marks_analyser(stud_marks):
    marks=list(stud_marks.values())
    average=st.mean(marks)
    minm=min(marks)
    maxm=max(marks)
    print(f"\n### The average mark is {average}")
    count=0
    toppr=[]
    low=[]
    avg=[]
    for k in stud_marks:
        if stud_marks[k]==maxm:
            toppr.append(k)
        elif stud_marks[k]==minm:
            low.append(k)   
        elif abs(stud_marks[k]-average)<=1:
            count+=1
            avg.append(k)
    print(f"\n{len(toppr)} students have scored {maxm} names are {' ,'.join(s.upper() for s in toppr)}")
    print(f"\n{len(low)} students have scored {minm} names are {' ,'.join(s.upper() for s in low)}")
    print(f"\n{len(avg)} students have scored {average}±1 names are {' ,'.join(s.upper() for s in avg)}")
stud_marks={
    'sumit':18,'krk':19,'kks':15,'rajiv': 16,'priya': 20,'amit': 14,'deepa': 17,'vivek': 19,'sana': 18,'mukesh': 12,'neha': 15,'rahul': 11,'shilpa': 20,
}
"""print("Enter name and marks of student.\n***Enter stop and 0 to stop entering***")
while (True):
    key=input("Enter name of student")
    if key.lower()=="stop":
        print("Stopped")
        break
    value=int(input(f"Enter marks of {key}:"))
    stud_marks[key]=value"""

#report printing
if stud_marks:    
    print("***The Report as follows :****")
    marks_analyser(stud_marks)
    data = pd.Series(stud_marks, name='Marks').to_frame()#data frame
    print(data)
    marks=pd.Series(stud_marks.values(),name='Marks')
    print(marks.describe())
else:
    print("Please enter data!!!")
```
